```python
import itertools

 
                   # FUNCTION TO CHECK IF ONE OF THE PLAYERS HAS WON
def winning(khel):      
   
    def all_items_same(check_l):
        if check_l.count(check_l[0]) == len(check_l)and check_l[0] != 0:
            print(f' Congrats to you : player {check_l[0]}')
            return True
            
    wingame = 0
    
        #for a horizontal win
    countrow= 0
    for row in khel:
        #print(row)
        countrow += 1
        if all_items_same(row):                  # Similar to "row.count(row[0]) == len(row) and row[0] != 0:"
            print(f'Player won on row {countrow}')
            print(f'She is player {row[0]}')
            wingame = 1
           
            return True
         
          #for a vertical win
    for col in range(len(khel)):              
        check_list = []
        for row in khel:
            #print(row[col])
            check_list.append(row[col])
                  
        if all_items_same(check_list):       # Similar to 
                               # "check_list.count(check_list[0]) ==len(check_list) and check_list[0] !=0" 
            print(f'voila, player {row[col]} won at column {col}')
            wingame = 1
            return True
    
                 #for diagonal win - left to right
    check_list = []
    for point in range(len(khel)):
        check_list.append(khel[point][point])
                  
    if all_items_same(check_list):     #check_list.count(check_list[0]) ==len(check_list) and check_list[0]!=0
        print(f'player {check_list[0]} won left to right diagonally')
        wingame = 1
        return True
        
    #for diagonal win - right to left
    check_another = []
    for point in range(0, len(khel)):
        check_another.append(khel[point][len(khel) - point -1])    
                  
    if all_items_same(check_another):            
                        #check_anothr.count(check_anothr[0]) ==len(check_anothr) and check_anothr[0]!=0
        print(f'player {check_another[0]} won right to left diagonally')
        wingame = 1
        return True
    
    #result:
    return False
    
#-------------------

     # FUNCTION TO UPDATE THE DISPLAY GAME BOX EVERY TIME A PLAYER PLAYS THEIR TURN
    
def game_board(game_map, player=0, row =0, column = 0, just_display =False):
    try:
        if game_map[row][column] != 0 :
            print("the place is already taken. Choose another position")
            return (game_map, False)
        
         #printing table index for dynamic size
            
                     '''print("  0, 1, 2")
                        s = " "
                        for i in range(game_size):
                            s += "  " + str(i)
                       '''
        #or
        s = '   ' + '  '.join([str(i) for i in range (game_size)])            
        print(s)
        
        
        #check if user has entered number or not
        if not just_display:
            game_map[row][column] = player
        for count, row in enumerate(game_map):
            print(count, row)
        return (game_map, True)
    
    except IndexError as e:
        print("Error : you should enter correct column number, not beyond limits", e)
        return (game_map, False)
    except Exception as e:
        print("something  went very wrong", e)
        return (game_map, False)
    
    
   #----------------

                # START THE APPLICATION. 
play = True
players = [1, 2]

while play:     # You can play game multiple times. Only when you don't want to, it will exit.
    
             # Create a dynamic sized box/matrix
    game_size = int(input('What size you want for tic tac toe\n '))     
                               
                    '''
                        khel = []
                        for i in range(game_size):
                            row = []
                            for k in range(game_size):
                                row.append(0)
                            game.append(row)
                            '''
      #or
    khel = [[0 for i in range(game_size)] for k in range(game_size)]
    
    
    
                 # START THE NEW INSTANCE OF THE GAME
    wingame = 0
    khel, _ = game_board(khel, just_display = True)     #second returning value is not to be cared. so '_'.
                             #every game should be started with this matrix(of zeroes) before the play begins
        
    player_choice = itertools.cycle([1,2])
    
    while not wingame:          #take inputs and all until there is a winner
        
        current_player = next(player_choice)
        print(f'current player is {current_player}')
        played = False
        
        while not played:      #ask him to enter number, until he enters correct value.
            column_choice = int(input("what column to enter number into  "))     
                                                      #input is taken as a string, so we need to
                                                    #convert it into an 'integer'
            row_choice = int(input("what row to enter number into  "))                
            
            khel, played = game_board(khel, current_player, row_choice, column_choice)
            
        if winning(khel):             #check for result after every entry is made
            wingame = True
            again = input('The game is over. Do you want to play it again? y/n')
            if again.lower() == 'y':
                print('restarting')
                pass
            elif again.lower() == 'n':
                print('Bye bye :) ')
                play = False
            else:
                print('not a valid anwer. So bye')
                play = False
```
